```python
import math

# Square
s = int(input("Square side: "))
print("Area:", s * s)

# Rectangle
l = int(input("Rectangle length: "))
w = int(input("Rectangle width: "))
print("Area:", l * w)

# Circle
r = int(input("Circle radius: "))
print("Area:", round(math.pi * r * r, 2))

# Triangle
b = int(input("Triangle base: "))
h = int(input("Triangle height: "))
print("Area:", 0.5 * b * h)
```

```python
import math

# Parallelogram
b = int(input("Parallelogram base: "))
h = int(input("Parallelogram height: "))
print("Area:", b * h)

# Trapezoid
a = int(input("Trapezoid top side: "))
b = int(input("Trapezoid bottom side: "))
h = int(input("Trapezoid height: "))
print("Area:", 0.5 * (a + b) * h)

# Rhombus
d1 = int(input("Rhombus diagonal 1: "))
d2 = int(input("Rhombus diagonal 2: "))
print("Area:", 0.5 * d1 * d2)

# Regular Pentagon
s = int(input("Pentagon side: "))
# Area = 1/4 * sqrt(5(5+2sqrt(5))) * s^2
print("Area:", round(1.720477 * s * s, 2))

# Regular Hexagon
s = int(input("Hexagon side: "))
# Area = (3*sqrt(3)/2) * s^2
print("Area:", round(2.598 * s * s, 2))

# Ellipse
a = int(input("Ellipse major axis (a): "))
b = int(input("Ellipse minor axis (b): "))
print("Area:", round(math.pi * a * b, 2))

# Sector of a Circle
r = int(input("Sector radius: "))
d = int(input("Sector angle in degrees: "))
print("Area:", round((d / 360) * math.pi * r * r, 2))
```
