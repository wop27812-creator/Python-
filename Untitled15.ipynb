# Mathematical Formula Calculator

This program will prompt you to input values for various mathematical formulas and then calculate the result. You can add more formulas by following the existing patterns.

```python
import math

def calculate_area_circle():
    """Calculates the area of a circle."""
    print("\n--- Area of a Circle ---")
    try:
        radius = float(input("Enter the radius of the circle: "))
        if radius < 0:
            print("Radius cannot be negative.")
            return
        area = math.pi * (radius ** 2)
        print(f"The area of the circle with radius {radius} is: {area:.2f}")
    except ValueError:
        print("Invalid input. Please enter a number.")

def calculate_area_rectangle():
    """Calculates the area of a rectangle."""
    print("\n--- Area of a Rectangle ---")
    try:
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        if length < 0 or width < 0:
            print("Length and width cannot be negative.")
            return
        area = length * width
        print(f"The area of the rectangle with length {length} and width {width} is: {area:.2f}")
    except ValueError:
        print("Invalid input. Please enter a number.")

def calculate_simple_interest():
    """Calculates simple interest."""
    print("\n--- Simple Interest ---")
    try:
        principal = float(input("Enter the principal amount: "))
        rate = float(input("Enter the annual interest rate (as a decimal, e.g., 0.05 for 5%): "))
        time = float(input("Enter the time in years: "))
        if principal < 0 or rate < 0 or time < 0:
            print("Principal, rate, and time cannot be negative.")
            return
        interest = principal * rate * time
        print(f"The simple interest is: {interest:.2f}")
    except ValueError:
        print("Invalid input. Please enter a number.")

def calculate_volume_cube():
    """Calculates the volume of a cube."""
    print("\n--- Volume of a Cube ---")
    try:
        side = float(input("Enter the side length of the cube: "))
        if side < 0:
            print("Side length cannot be negative.")
            return
        volume = side ** 3
        print(f"The volume of the cube with side {side} is: {volume:.2f}")
    except ValueError:
        print("Invalid input. Please enter a number.")

def convert_fahrenheit_to_celsius():
    """Converts Fahrenheit to Celsius."""
    print("\n--- Fahrenheit to Celsius Conversion ---")
    try:
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit}°F is equal to {celsius:.2f}°C")
    except ValueError:
        print("Invalid input. Please enter a number.")

# You can define more functions for other formulas here.
# For example:
# def calculate_quadratic_roots():
#     print("\n--- Quadratic Roots ---")
#     try:
#         a = float(input("Enter coefficient a: "))
#         b = float(input("Enter coefficient b: "))
#         c = float(input("Enter coefficient c: "))
#         # Add quadratic formula logic here
#     except ValueError:
#         print("Invalid input. Please enter a number.")


def main():
    """Main function to run all formula calculations."""
    print("Welcome to the Mathematical Formula Calculator!")

    # Call each formula function
    calculate_area_circle()
    calculate_area_rectangle()
    calculate_simple_interest()
    calculate_volume_cube()
    convert_fahrenheit_to_celsius()
    # Call your new formula functions here, e.g.:
    # calculate_quadratic_roots()

    print("\nThank you for using the calculator!")

if __name__ == "__main__":
    main()
```
